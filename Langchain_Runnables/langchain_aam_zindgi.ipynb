```python
import random
```

```python
class NakliLLM:

    def __init__(self):
        print("LLM initialized")

    def predict(self, prompt):

        response_list = [
            "AI is the creation of human intelligence in machines.",
            "Machine learning is a subset of AI that focuses on learning from data.",
            "Deep learning is a type of machine learning that uses neural networks.",
            "Natural language processing enables machines to understand human language.",
            "Computer vision allows machines to interpret visual information."
        ]

        return {'response': random.choice(response_list)}
```

```python
llm = NakliLLM()
```

```python
llm.predict("What is AI?")
```

```python
class NakliPromptTemplate:

    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate("Explain the concept of {concept} in {length} terms.", 
                               input_variables=["concept", "length"])
```

```python
prompt =template.format({'concept': "Artificial Intelligence", 'length': "simple"})
```

```python
llm = NakliLLM()
```

```python
llm.predict(prompt)
```

```python
class NakliLLMChain:

    def __init__(self, llm, prompt):
        self.llm = llm
        self.prompt = prompt
    
    def run(self, input_dict):

        formatted_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(formatted_prompt)

        return result['response']
```

```python
prompt = template.format({'concept': "Artificial Intelligence", 'length': "simple"})
```

```python
llm = NakliLLM()
```

```python
chain = NakliLLMChain(llm=llm, prompt=template)
```

```python
chain.run({'concept': "AI", 'length': "simple"})
```
